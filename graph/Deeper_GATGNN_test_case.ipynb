```python
import os
from pymatgen.io.cif import CifParser
from pymatgen.io.vasp.inputs import Poscar

# Specify the folder path containing the CIF files
folder_path = "/content/test11"

# Create the target folder to store the VASP files
target_folder = "mm"
os.makedirs(target_folder, exist_ok=True)

# Convert CIF files to VASP files
for filename in os.listdir(folder_path):
    if filename.endswith(".cif"):
        # Read the CIF file
        cif_filepath = os.path.join(folder_path, filename)
        cif_parser = CifParser(cif_filepath)
        structure = cif_parser.get_structures(primitive=False)[0]

        # Write the structure to a VASP (POSCAR) file
        vasp_filename = os.path.splitext(filename)[0] + ".vasp"
        vasp_filepath = os.path.join(target_folder, vasp_filename)
        poscar = Poscar(structure)
        poscar.write_file(vasp_filepath)
```

```python
import os
import pandas as pd

# Path to the directory containing CIF files and the CSV file
directory_path = '/content/test'
csv_file_path = os.path.join(directory_path, 'targets11.csv')  # Update this with your CSV file name

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# List all CIF files in the directory
cif_files = [f for f in os.listdir(directory_path) if f.endswith('.cif')]

# Create a set of material IDs that have a corresponding CIF file
material_ids_with_cif = set([filename.replace('.cif', '') for filename in cif_files])

# Filter the DataFrame to keep only rows where the material ID has a corresponding CIF file
# Assuming the first column contains the material IDs
filtered_df = df[df[df.columns[0]].isin(material_ids_with_cif)]

# Save the filtered DataFrame to a new CSV file
filtered_csv_file_path = os.path.join(directory_path, 'targets.csv')
filtered_df.to_csv(filtered_csv_file_path, index=False)

print(f'Filtered CSV file saved to: {filtered_csv_file_path}')
```
