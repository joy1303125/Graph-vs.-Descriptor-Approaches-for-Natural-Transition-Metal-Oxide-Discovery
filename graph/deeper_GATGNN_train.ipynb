```python
import os

# The directory containing your POSCAR files
directory = '/content/input'  # Replace with the path to your directory

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.startswith('POSCAR-JVASP-') and filename.endswith('.vasp'):
        # Extract the ID part of the filename
        id_part = filename.split('-')[2].split('.')[0]

        # Define the new filename format
        new_filename = f'mp-{id_part}.vasp'

        # Define the old and new file paths
        old_file = os.path.join(directory, filename)
        new_file = os.path.join(directory, new_filename)

        # Rename the file
        os.rename(old_file, new_file)
        print(f'Renamed {old_file} to {new_file}')
```
